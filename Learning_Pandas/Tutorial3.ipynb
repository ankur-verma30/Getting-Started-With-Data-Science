```python
# Method 3 to create a dataframe using the dictionary
# df=pd.DataFrame({passing the name of dictionary})
weather_dict_data={
    'day':['1/1/2017','1/2/2017','1/3/2017','1/4/2017'],
    'temperature':[1,2,3,4],
    'windspeed':[6,7,8,9],
    'event':['Rain','Sunny','Snow','Windy']
}
df3=pd.DataFrame(weather_dict_data)
```
